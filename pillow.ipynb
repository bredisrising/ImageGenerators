```python
from PIL import Image
import numpy as np
```

```python
frame = Image.open(f"./results/{74}timestep_diffused.png")
```

```python
frame = np.array(frame)
```

```python
frame.shape
```

```python
a = frame[:, :, 0]
b = frame[:, :, 1]
c = frame[:, :, 2]

a = np.kron(a, np.ones((25, 25)))
b = np.kron(b, np.ones((25, 25)))
c = np.kron(c, np.ones((25, 25)))
```

```python
a = np.expand_dims(a, axis=-1)
b = np.expand_dims(b, axis=-1)
c = np.expand_dims(c, axis=-1)
```

```python
bigger = np.concatenate((a, b, c), axis=2)
bigger.shape
```

```python
bigger = bigger.astype(np.uint8)
```
