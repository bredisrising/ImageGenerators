```python
import torch
import matplotlib.pyplot as plt
```

```python
normal_noise = torch.randn((64, 64, 3))
plt.imshow(normal_noise)
```

```python
normal_noise.mean(), normal_noise.var()
```

```python
mu = 1
var = 2

new_noise = normal_noise * var + mu
plt.imshow(new_noise)
print(new_noise.mean(), new_noise.var())
```
